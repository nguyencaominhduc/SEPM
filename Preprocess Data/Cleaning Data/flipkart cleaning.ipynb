```python
fixed=flipkart[['product_name','retail_price','brand','product_url','image']]
```

```python
fixed=fixed.dropna()
```

```python
index=0
for row in fixed['image']:
    newurl = row.split('"')[1].split('"')[0]
    fixed['image'][index]=newurl
    index+=1  
#     print(index)   
```
